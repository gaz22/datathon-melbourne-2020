```python
import numpy as np
import pandas as pd
import glob, os    
import matplotlib.pyplot as plt
```

```python
# 2020 data
electricity_data_2020 = pd.read_csv("dataset/electricity_data_2020.csv", index_col=None, header=0)
```

```python
electricity_data_2020['SETTLEMENTDATE'] = pd.to_datetime(electricity_data_2020['SETTLEMENTDATE'])
electricity_data_2020 = electricity_data_2020[electricity_data_2020['SETTLEMENTDATE'] < '2020-09-01']
electricity_data_2020.sort_values(by='SETTLEMENTDATE')
```

```python
electricity_data_2020_mean = electricity_data_2020.groupby(pd.Grouper(key='SETTLEMENTDATE', freq='1H')).mean()
electricity_data_2020_mean = electricity_data_2020_mean[electricity_data_2020_mean.index.dayofweek < 5]
electricity_data_2020_mean
```

```python
electricity_data_2020_mean = electricity_data_2020_mean[(electricity_data_2020_mean.index.date.astype(str) != "2020-01-01") & (electricity_data_2020_mean.index.date.astype(str) != "2020-01-27") & (electricity_data_2020_mean.index.date.astype(str) != "2020-03-09") & (electricity_data_2020_mean.index.date.astype(str) != "2020-04-10") & (electricity_data_2020_mean.index.date.astype(str) != "2020-04-13") & (electricity_data_2020_mean.index.date.astype(str) != "2020-04-25") & (electricity_data_2020_mean.index.date.astype(str) != "2020-06-08")]
electricity_data_2020_mean
```

```python
electricity_data_2020_mean.index = electricity_data_2020_mean.index.time
electricity_data_2020_mean.index.names = ['SETTLEMENTDATE']
electricity_data_2020_mean
```

```python
electricity_data_2020_max = electricity_data_2020.groupby(pd.Grouper(key='SETTLEMENTDATE', freq='1H')).max()
electricity_data_2020_max = electricity_data_2020_max[electricity_data_2020_max.index.dayofweek < 5]

electricity_data_2020_max = electricity_data_2020_max[(electricity_data_2020_max.index.date.astype(str) != "2020-01-01") & (electricity_data_2020_max.index.date.astype(str) != "2020-01-27") & (electricity_data_2020_max.index.date.astype(str) != "2020-03-09") & (electricity_data_2020_max.index.date.astype(str) != "2020-04-10") & (electricity_data_2020_max.index.date.astype(str) != "2020-04-13") & (electricity_data_2020_max.index.date.astype(str) != "2020-04-25") & (electricity_data_2020_max.index.date.astype(str) != "2020-06-08")]

electricity_data_2020_max.index = electricity_data_2020_max.index.time
electricity_data_2020_max.index.names = ['SETTLEMENTDATE']
electricity_data_2020_max
```

```python
electricity_data_2020_min = electricity_data_2020.groupby(pd.Grouper(key='SETTLEMENTDATE', freq='1H')).min()
electricity_data_2020_min = electricity_data_2020_min[electricity_data_2020_min.index.dayofweek < 5]

electricity_data_2020_min = electricity_data_2020_min[(electricity_data_2020_min.index.date.astype(str) != "2020-01-01") & (electricity_data_2020_min.index.date.astype(str) != "2020-01-27") & (electricity_data_2020_min.index.date.astype(str) != "2020-03-09") & (electricity_data_2020_min.index.date.astype(str) != "2020-04-10") & (electricity_data_2020_min.index.date.astype(str) != "2020-04-13") & (electricity_data_2020_min.index.date.astype(str) != "2020-04-25") & (electricity_data_2020_min.index.date.astype(str) != "2020-06-08")]

electricity_data_2020_min.index = electricity_data_2020_min.index.time
electricity_data_2020_min.index.names = ['SETTLEMENTDATE']
electricity_data_2020_min
```

```python
electricity_data_2020_mean['TOTALDEMAND_MIN'] = electricity_data_2020_min['TOTALDEMAND']
electricity_data_2020_mean['TOTALDEMAND_MAX'] = electricity_data_2020_max['TOTALDEMAND']
electricity_data_2020_mean
```

```python
electricity_data_2020_mean.to_csv('./compiled_data/electricity_data_2020_weekdays_mean_max_min.csv')
```
