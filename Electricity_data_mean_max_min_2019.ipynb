```python
import numpy as np
import pandas as pd
import glob, os    
import matplotlib.pyplot as plt
```

```python
# 2019 data
electricity_data_2019 = pd.read_csv("dataset/electricity_data_2019.csv", index_col=None, header=0)
```

```python
electricity_data_2019['SETTLEMENTDATE'] = pd.to_datetime(electricity_data_2019['SETTLEMENTDATE'])
electricity_data_2019 = electricity_data_2019[electricity_data_2019['SETTLEMENTDATE'] < '2019-09-01']
electricity_data_2019.sort_values(by='SETTLEMENTDATE')
```

```python
electricity_data_2019_mean = electricity_data_2019.groupby(pd.Grouper(key='SETTLEMENTDATE', freq='1H')).mean()
electricity_data_2019_mean = electricity_data_2019_mean[electricity_data_2019_mean.index.dayofweek < 5]
electricity_data_2019_mean
```

```python
electricity_data_2019_mean = electricity_data_2019_mean[(electricity_data_2019_mean.index.date.astype(str) != "2019-01-01") & (electricity_data_2019_mean.index.date.astype(str) != "2019-01-28") & (electricity_data_2019_mean.index.date.astype(str) != "2019-03-11") & (electricity_data_2019_mean.index.date.astype(str) != "2019-04-19") & (electricity_data_2019_mean.index.date.astype(str) != "2019-04-22") & (electricity_data_2019_mean.index.date.astype(str) != "2019-04-25") & (electricity_data_2019_mean.index.date.astype(str) != "2019-06-10")]
electricity_data_2019_mean
```

```python
electricity_data_2019_mean.index = electricity_data_2019_mean.index.time
electricity_data_2019_mean.index.names = ['SETTLEMENTDATE']
electricity_data_2019_mean

```

```python
electricity_data_2019_mean[(electricity_data_2019_mean.index.astype(str) == "16:00:00")]['TOTALDEMAND'].min()
```

```python
electricity_data_2019_mean = electricity_data_2019.groupby(pd.Grouper(key='SETTLEMENTDATE', freq='1H')).std()
electricity_data_2019_mean
```

```python
electricity_data_2019_max = electricity_data_2019.groupby(pd.Grouper(key='SETTLEMENTDATE', freq='1H')).max()
electricity_data_2019_max = electricity_data_2019_max[electricity_data_2019_max.index.dayofweek < 5]

electricity_data_2019_max = electricity_data_2019_max[(electricity_data_2019_max.index.date.astype(str) != "2019-01-01") & (electricity_data_2019_max.index.date.astype(str) != "2019-01-28") & (electricity_data_2019_max.index.date.astype(str) != "2019-03-11") & (electricity_data_2019_max.index.date.astype(str) != "2019-04-19") & (electricity_data_2019_max.index.date.astype(str) != "2019-04-22") & (electricity_data_2019_max.index.date.astype(str) != "2019-04-25") & (electricity_data_2019_max.index.date.astype(str) != "2019-06-10")]

electricity_data_2019_max.index = electricity_data_2019_max.index.time
electricity_data_2019_max.index.names = ['SETTLEMENTDATE']
electricity_data_2019_max
```

```python
electricity_data_2019_min = electricity_data_2019.groupby(pd.Grouper(key='SETTLEMENTDATE', freq='1H')).min()
electricity_data_2019_min = electricity_data_2019_min[electricity_data_2019_min.index.dayofweek < 5]

electricity_data_2019_min = electricity_data_2019_min[(electricity_data_2019_min.index.date.astype(str) != "2019-01-01") & (electricity_data_2019_min.index.date.astype(str) != "2019-01-28") & (electricity_data_2019_min.index.date.astype(str) != "2019-03-11") & (electricity_data_2019_min.index.date.astype(str) != "2019-04-19") & (electricity_data_2019_min.index.date.astype(str) != "2019-04-22") & (electricity_data_2019_min.index.date.astype(str) != "2019-04-25") & (electricity_data_2019_min.index.date.astype(str) != "2019-06-10")]

electricity_data_2019_min.index = electricity_data_2019_min.index.time
electricity_data_2019_min.index.names = ['SETTLEMENTDATE']
electricity_data_2019_min
```

```python
electricity_data_2019_mean['TOTALDEMAND_MIN'] = electricity_data_2019_min['TOTALDEMAND']
electricity_data_2019_mean['TOTALDEMAND_MAX'] = electricity_data_2019_max['TOTALDEMAND']
```

```python
electricity_data_2019_mean
```

```python
electricity_data_2019_mean.to_csv('./compiled_data/2019_weekdays/electricity_data_2019_weekdays_mean_max_min.csv')
```
